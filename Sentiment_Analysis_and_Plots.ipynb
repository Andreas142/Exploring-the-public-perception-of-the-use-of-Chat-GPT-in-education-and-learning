```python
#Import the required libraries
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
#Download the necessary NLTK resources
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
#Initialize the sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
#Apply tokenization, stop word removal, lemmatization, and noise removal
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

filtered_data_education['tokens'] = filtered_data_education['content'].apply(word_tokenize)
filtered_data_education['tokens'] = filtered_data_education['tokens'].apply(lambda x: [word for word in x if word.lower() not in stop_words])
filtered_data_education['tokens'] = filtered_data_education['tokens'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
filtered_data_education['content'] = filtered_data_education['content'].apply(lambda x: re.sub(r'http\S+|www\.\S+|[^a-zA-Z\s]', '', x))
```

```python
#Apply the sentiment analysis using VADER and categorize the sentiment scores
filtered_data_education['sentiment'] = filtered_data_education['content'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

def categorize_sentiment(score):
    if score > 0:
        return 'positive'
    elif score < 0:
        return 'negative'
    else:
        return 'neutral'

filtered_data_education['sentiment_category'] = filtered_data_education['sentiment'].apply(categorize_sentiment)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate sentiment category counts
sentiment_counts = filtered_data_education['sentiment_category'].value_counts()

# Define colors for each sentiment category
colors = ['blue', 'yellow', 'red']

# Create a bar chart
plt.figure(figsize=(10, 9))
plt.bar(sentiment_counts.index, sentiment_counts, color=colors)

# Add labels and title
plt.xlabel('Sentiment Category', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Sentiment Analysis Results', fontsize=16)

# Calculate total count of data points
total_count = len(filtered_data_education)

# Create the text to display on the bar chart
text = f"Total Count: {total_count}\n"
text += '\n'.join([f"{category.capitalize()}: {count} ({(count / total_count * 100):.1f}%)" for category, count in sentiment_counts.items()])

# Add the text as an annotation above the bars
for i, count in enumerate(sentiment_counts):
    plt.text(i, count + 5, f'{count}\n({(count / total_count * 100):.1f}%)', ha='center', fontsize=12, color='black')

# Customize the appearance
plt.xticks(fontsize=12, rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.yticks(fontsize=12)  # Customize y-axis tick font size
plt.tight_layout()  # Adjust spacing to prevent overlapping

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dictionary to store the counts for each category
category_counts = {
    'education ': 0,
    'academic': 0,
    'teaching': 0,
    'learning': 0,
    'school': 0,
    'classroom': 0,
    'teachers': 0,
    'students': 0,
    'online learning': 0,
    'e-learning': 0,
    'educational': 0,
    'teacher ': 0,
    'teach ': 0,
    'learn ': 0,
    'academia': 0,
    'university': 0,
    'college': 0,
    'educate': 0,
    'student ': 0,
    'pupil': 0,
    'lecturer': 0,
    'professor': 0,
    'tutor': 0
}


# Iterate over each row in the DataFrame
for index, row in filtered_data_education.iterrows():
    tweet_content = row['content']
    # Check if the tweet contains any of the keywords
    for category in category_counts:
        if category in tweet_content:
            category_counts[category] += 1

# Convert the counts to a pandas DataFrame
counts_df = pd.DataFrame.from_dict(category_counts, orient='index', columns=['Count'])

# Calculate the percentages
counts_df['Percentage'] = counts_df['Count'] / len(filtered_data_education) * 100

# Sort the DataFrame by count in descending order
counts_df = counts_df.sort_values('Count', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Percentage', y=counts_df.index, data=counts_df, color='purple')
plt.xlabel('Percentage')
plt.ylabel('Category')
plt.title('Tweet Percentages for Each Category')
plt.tight_layout()
plt.show()

```

```python

# Define the keywords
keywords = ['education ', 'learn ', 'learning', 'students', 'school', 'teachers']

# Initialize empty lists to store the sentiment category percentages
positive_percentages = []
neutral_percentages = []
negative_percentages = []

# Calculate the sentiment category percentages for each keyword
for keyword in keywords:
    # Filter the DataFrame based on the keyword
    filtered_data_keywords = filtered_data_education[filtered_data_education['content'].str.contains(keyword, case=False)]

    # Calculate the sentiment category counts for the keyword
    sentiment_counts_by_keyword = filtered_data_keywords['sentiment_category'].value_counts(normalize=True) * 100

    # Append the percentages to the corresponding lists
    positive_percentages.append(sentiment_counts_by_keyword.get('positive', 0))
    neutral_percentages.append(sentiment_counts_by_keyword.get('neutral', 0))
    negative_percentages.append(sentiment_counts_by_keyword.get('negative', 0))

# Create a DataFrame from the sentiment category percentages
data = pd.DataFrame({
    'Sentiment Category': ['Positive', 'Neutral', 'Negative'],
    'Education': [positive_percentages[0], neutral_percentages[0], negative_percentages[0]],
    'Learn': [positive_percentages[1], neutral_percentages[1], negative_percentages[1]],
    'Learning': [positive_percentages[2], neutral_percentages[2], negative_percentages[2]],
    'Students': [positive_percentages[3], neutral_percentages[3], negative_percentages[3]],
    'School': [positive_percentages[4], neutral_percentages[4], negative_percentages[4]],
    'Teachers': [positive_percentages[5], neutral_percentages[5], negative_percentages[5]]
})

# Plot the data
plt.figure(figsize=(10, 6))
bar_width = 0.12
index = data.index
colors = ['deepskyblue', 'skyblue', 'lightslategrey', 'olivedrab', 'darkkhaki', 'palegoldenrod']

plt.bar(index, data['Education'], bar_width, label='Education', color=colors[0])
plt.bar(index + bar_width, data['Learn'], bar_width, label='Learn', color=colors[1])
plt.bar(index + 2 * bar_width, data['Learning'], bar_width, label='Learning', color=colors[2])
plt.bar(index + 3 * bar_width, data['Students'], bar_width, label='Students', color=colors[3])
plt.bar(index + 4 * bar_width, data['School'], bar_width, label='School', color=colors[4])
plt.bar(index + 5 * bar_width, data['Teachers'], bar_width, label='Teachers', color=colors[5])

plt.xlabel('Sentiment Category')
plt.ylabel('Percentage')
plt.title('Sentiment Analysis')
plt.xticks(index + (bar_width * (len(keywords) - 1)) / 2, data['Sentiment Category'])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.stats import linregress

# Set plot style and background color
sns.set(style='whitegrid', rc={'axes.facecolor': 'white', 'grid.color': 'grey'})  # Set the background color to white with black grid lines
plt.figure(facecolor='white')  # Set the background color to white

# Create a figure and axes
fig, ax = plt.subplots(figsize=(15, 7))

# Scatter plot the sentiment counts for each category
sns.scatterplot(x=pd.to_numeric(sentiment_counts.index), y='negative', data=sentiment_counts, color='red', ax=ax)
sns.scatterplot(x=pd.to_numeric(sentiment_counts.index), y='neutral', data=sentiment_counts, color='yellow', ax=ax)
sns.scatterplot(x=pd.to_numeric(sentiment_counts.index), y='positive', data=sentiment_counts, color='blue', ax=ax)

# Calculate lowess regression lines for each sentiment category
sns.regplot(x=pd.to_numeric(sentiment_counts.index), y='negative', data=sentiment_counts, scatter=False, lowess=True, color='red', ax=ax)
sns.regplot(x=pd.to_numeric(sentiment_counts.index), y='neutral', data=sentiment_counts, scatter=False, lowess=True, color='yellow', ax=ax)
sns.regplot(x=pd.to_numeric(sentiment_counts.index), y='positive', data=sentiment_counts, scatter=False, lowess=True, color='blue', ax=ax)

# Set axis labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Count')
ax.set_title('Sentiment Analysis over Time')

# Calculate and print R-squared values
x_values = pd.to_numeric(sentiment_counts.index)
y_values = sentiment_counts['negative']
negative_slope, negative_intercept, negative_rvalue, _, _ = linregress(x_values, y_values)
negative_r2 = negative_rvalue ** 2

y_values = sentiment_counts['neutral']
neutral_slope, neutral_intercept, neutral_rvalue, _, _ = linregress(x_values, y_values)
neutral_r2 = neutral_rvalue ** 2

y_values = sentiment_counts['positive']
positive_slope, positive_intercept, positive_rvalue, _, _ = linregress(x_values, y_values)
positive_r2 = positive_rvalue ** 2

print(f"R-squared values: Negative={negative_r2}, Neutral={neutral_r2}, Positive={positive_r2}")

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statsmodels.api as sm

# Set plot style and background color
sns.set(style='whitegrid', rc={'axes.facecolor': 'white', 'grid.color': 'grey'})
plt.figure(facecolor='white')

# Create a figure and axes
fig, ax = plt.subplots(figsize=(15, 7))

# Scatter plot the sentiment counts for each category
sns.scatterplot(x=sentiment_counts.index, y='negative', data=sentiment_counts, color='red', ax=ax, label='Negative')
sns.scatterplot(x=sentiment_counts.index, y='neutral', data=sentiment_counts, color='yellow', ax=ax, label='Neutral')
sns.scatterplot(x=sentiment_counts.index, y='positive', data=sentiment_counts, color='blue', ax=ax, label='Positive')

# Calculate lowess regression lines using statsmodels
lowess = sm.nonparametric.lowess
lowess_negative = lowess(sentiment_counts['negative'], sentiment_counts.index, frac=0.2)
lowess_neutral = lowess(sentiment_counts['neutral'], sentiment_counts.index, frac=0.2)
lowess_positive = lowess(sentiment_counts['positive'], sentiment_counts.index, frac=0.2)

# Plot lowess regression lines
ax.plot(sentiment_counts.index, lowess_negative[:, 1], color='red', label='Negative (Lowess)')
ax.plot(sentiment_counts.index, lowess_neutral[:, 1], color='yellow', label='Neutral (Lowess)')
ax.plot(sentiment_counts.index, lowess_positive[:, 1], color='blue', label='Positive (Lowess)')

# Set axis labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Count')
ax.set_title('Sentiment Analysis over Time')

# Show every 3 days' dates on the x-axis
plt.xticks(sentiment_counts.index[::3], sentiment_counts.index.strftime('%Y-%m-%d')[::3], rotation=45)

plt.tight_layout()
plt.legend()
plt.show()
```
